## Append CSV files

```python
files = ['data_analyst_ca.csv', 'data_scientist_ca.csv', 'data_engineer_ca.csv', 'python_developer_ca.csv']
```

```python
df_csv_append = pd.DataFrame()
 
# append the CSV files
for file in files:
    df = pd.read_csv(file, header=None)
    df[file] = True
    df_csv_append = pd.concat([df_csv_append, df])
 
df_csv_append.head()
```

```python
len(df_csv_append)
```

```python
df = df_csv_append.drop_duplicates()
```

```python
total_jobs = len(df)
```

```python
df.columns = ['Job Description', 'Data Analyst', 'Data Scientist', 'Data Engineer', 'Python Developer']
df['Job Description'] = df['Job Description'].str.lower()
```

```python
#df['Certification Required?'] = df['Job Description'].str.contains('exam') 
df['Certification Required?'] = df['Job Description'].str.contains('exam') | df['Job Description'].str.contains('certif')
```

```python
df = df.fillna(False).reset_index(drop=True)
df = df.drop(columns='Job Description')
```

```python
df.to_csv('JobDescriptionDataset.csv', index=False)
```
